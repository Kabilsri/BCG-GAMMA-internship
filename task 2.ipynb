```python
# Load necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
client_data = pd.read_csv("client_data.csv")
price_data = pd.read_csv("price_data.csv")

# Display dataset info (data types of each column)
print("Client Data Types:")
print(client_data.dtypes)

print("\nPrice Data Types:")
print(price_data.dtypes)

# Display first few rows
display(client_data.head())
display(price_data.head())
```

```python
# Descriptive statistics
print("Client Data Descriptive Statistics:")
display(client_data.describe())

print("Price Data Descriptive Statistics:")
display(price_data.describe())

# Checking unique values for categorical columns
categorical_columns = client_data.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"Unique values in {col}: {client_data[col].nunique()}")
```

```python
# Visualizing distributions of numeric columns
numeric_cols = client_data.select_dtypes(include=['int64', 'float64']).columns

# Set up the figure
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols[:6]):  # Limit to 6 for clarity
    plt.subplot(2, 3, i + 1)
    sns.histplot(client_data[col], bins=50, kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
# Visualizing distributions for price data
numeric_cols_price = price_data.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(15, 8))
for i, col in enumerate(numeric_cols_price[:6]):  # Limit to 6 for clarity
    plt.subplot(2, 3, i + 1)
    sns.histplot(price_data[col], bins=50, kde=True)
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```
