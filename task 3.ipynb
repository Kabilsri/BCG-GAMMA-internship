### Difference between off-peak prices in December and preceding January

Below is the code created by your colleague to calculate the feature described above. Use this code to re-create this feature and then think about ways to build on this feature to create features with a higher predictive power.

```python
price_df = pd.read_csv('price_data.csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head()
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
diff.head()
```

```python
# Load necessary libraries
import pandas as pd
import numpy as np

# Load the cleaned dataset after EDA
data = pd.read_csv("clean_data_after_eda.csv")

# Display initial dataset info
print("Initial Data Info:")
print(data.info())

# Drop columns with a single unique value or irrelevant columns
columns_to_drop = [col for col in data.columns if data[col].nunique() == 1]
data.drop(columns=columns_to_drop, inplace=True)
print(f"Removed columns: {columns_to_drop}")
```

```python
# Feature Engineering: Extracting date components
if 'signup_date' in data.columns:
    data['signup_year'] = pd.to_datetime(data['signup_date']).dt.year
    data['signup_month'] = pd.to_datetime(data['signup_date']).dt.month
    data['signup_day'] = pd.to_datetime(data['signup_date']).dt.day

# Creating new features based on consumption data
if 'forecast_cons_12m' in data.columns and 'cons_last_month' in data.columns:
    data['consumption_change_ratio'] = (data['cons_last_month'] / (data['forecast_cons_12m'] + 1e-5)).round(4)

# Log transformation for skewed data
skewed_cols = ['forecast_cons_12m', 'cons_last_month', 'net_margin']
for col in skewed_cols:
    if col in data.columns:
        data[f'log_{col}'] = np.log1p(data[col])
```

```python
# Save the transformed dataset
data.to_csv("final_feature_engineered_data.csv", index=False)
print("Feature engineering completed and saved.") 
```
